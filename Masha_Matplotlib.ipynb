```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```
# empty figure

fig, ax=plt.subplots()
```

```
x=[3,4,5,7,5,3,6]
y=[60,70,110,157,188,190,250]
fig, ax=plt.subplots()
ax.plot(y,x)
```

```
x=np.linspace(1,5,400)
y=x**2 #numpy broadcasting
fig, ax=plt.subplots()
ax.plot(y,x)
ax.legend()
```

```
sns.jointplot(x=np.linspace(1,5,400), y=x**2)
plt.show()
```
